```python
# Чтение данных из Excel
df = pd.read_excel('Координаты_Нижний_Новгород_New.xlsx')

# Инициализация словаря
polygons_R = {}

# Список названий цветов
color_names = [
    'red', 'blue', 'green', 'yellow', 'orange', 'purple', 'cyan', 'magenta',
    'lime', 'pink', 'brown', 'gray', 'olive', 'teal', 'navy', 'violet',
    'gold', 'silver', 'coral', 'indigo', 'salmon', 'plum', 'khaki', 'lavender'
]

# Перемешиваем список цветов, чтобы они были случайными
random.shuffle(color_names)

# Обработка каждой строки DataFrame
for index, row in df.iterrows():
    район = row['Район']
    координаты = ast.literal_eval(row['Координаты'])  # Преобразование строки в список кортежей
    color = color_names[index % len(color_names)]  # Получаем цвет из списка
    polygons_R[район] = [координаты, color]  # Добавляем координаты и цвет в словарь

# print(polygons_NN)
polygons_R
```

```python
def coordinates_from_link(link):
    """
    Преобразует ссылку на полигон с Яндекс Карт в список тюплов вида [(Д, Ш), (Д, Ш), ..., (Д, Ш)]
    
    link - столбец со ссылкой на полигон с Яндекс Карт (полученная при помощи инструмента Планиметр)
    """
    cut = re.search(r'rl=(.*?)&rlt', link)                                              #с помощью рег выражения вырезаем часть строки между 2 метками
    cut_link = cut.group(1)
    coordinates_list = cut_link.split('~')                                              #делим по '~' на координаты
    
    coordinates = []
    first_latitude, first_longitude = coordinates_list[0].split('%2C') 
    coordinates.append((float(first_longitude), float(first_latitude)))                 #получаем координату начальной точки полигона и фиксируем ее в списке
    
    previous_latitude = float(first_latitude)  
    previous_longitude = float(first_longitude)
    
    for i in range(1, len(coordinates_list)):
        delta_latitude, delta_longitude = coordinates_list[i].split('%2C')              #получаем значения дельты
        latitude = previous_latitude + float(delta_latitude)                            #получаем i-ое значение широты
        longitude = previous_longitude + float(delta_longitude)                         #получаем i-ое значение долготы
        latitude = round(latitude, 6)                                                   #фиксируем округление до 6 знаков после точки (иначе точка уплывает...)
        longitude = round(longitude, 6)
        coordinates.append((float(longitude), float(latitude)))                         #записываем кортеж в список
        previous_latitude = latitude                                                    #обновляем значения предыдущих широты и долготы
        previous_longitude = longitude
    
    return coordinates

def polygon_yandex(dataf, polygon_name, link, status):
    """
    Для каждой ссылки в dataf по строкам преобразует ссылку в координаты и создает новый df со столбцами 'Район' и 'Координаты'
    
    dataf - DataFrame, из которого используются данные 
    polygon_name - столбец с названием района
    link - столбец со ссылкой на полигон с Яндекс Карт (полученная при помощи инструмента Планиметр) 
    """
    coordinates = []
    for index, row in dataf.iterrows():                                                 #итерируем по строкам dataf
        extracted_coordinates = coordinates_from_link(row[link])                        #применяем 1 функцию
        coordinates.append(extracted_coordinates)
    df_1 = pd.DataFrame({'Район': dataf[polygon_name], 'Координаты': coordinates, 'Статус': dataf[status]})
    return df_1
```
